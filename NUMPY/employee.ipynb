```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('employee.csv')
```

```python
df.sample(5)
```

```python
df.describe
```

```python
df.info()
```

```python
# Checking the Missing Value

df.isnull().sum()
```

```python
# Filling missing Value in Salary

df['salary']=df['salary'].fillna(df['salary'].mean())
```

```python
df.isnull().sum()
```

```python
# df.drop_duplicates()
df.dropna(thresh=3)
df.isnull().sum()
```

```python
df['performance_rating']=df['performance_rating'].fillna(df['performance_rating'].median())
```

```python
df.isnull().sum()
```

```python
df['age']=df['age'].fillna(df['age'].median()).round().astype(int)
df['exp']=df['exp'].fillna(df['exp'].mean()).round().astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['name'] = df['name'].fillna('Unknown')
```

```python
df['city'] =df['city'].fillna(df['city'].mode()[0])
```

```python
df['dept'] =df['dept'].fillna(df['dept'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.to_csv('Cleaned_employee.csv',index=False) #We use index=False in to_csv() to prevent pandas from saving the DataFrame index as a separate column in the CSV file.
```
