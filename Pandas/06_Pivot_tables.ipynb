#### Pivot Tables
##### A pivot table takes a DataFrame and lets you:

##### Group data by one or more columns (index)

##### Aggregate values (values) using functions like sum, mean, count

##### Optionally create columns from unique values (columns)

```python
import numpy as np
import pandas as pd
```

```python
data={

    'Date' : pd.date_range('2026-01-01',periods=20),
    'Product' : ['A','B','C','D'] * 5,
    'Region' :['East','West','North','South','East','West','North','South','East','West','North','South','East','West','North','South','East','West','North','South'],
    'Sales':np.random.randint(100,1000,20),
    'Units':np.random.randint(10,100,20),
    'Rep'  : [
        'John','Mary','Bob','Alice','John','Mary','Bob','Alice','John','Mary','Bob','Alice',
        'John','Mary','Bob','Alice','John','Mary','Bob','Alice'
    ]

}
df=pd.DataFrame(data)
df['Months'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q'+df['Date'].dt.quarter.astype(str)

```

```python
df
```

```python
pd.pivot_table(df,values='Sales',index='Region',columns='Product')
```

```python
pd.pivot_table(df,values='Units',index='Product',columns='Region')
```

```python
pd.pivot_table(df,values=['Units','Sales'],index='Product',columns='Region')
```

```python
pd.pivot_table(df,values='Units',index='Product',columns='Region',aggfunc='sum')
```

#### Cross Tabs

```python
#cross tabulation (crosstab) is a way to compute a frequency table of two (or more) variables.

# It is like a pivot table, but specifically used to count occurrences and show relationships between categorical variables.
# 
```

```python
pd.crosstab(df['Region'],df['Product'])
```

```python
pd.crosstab(df['Region'],df['Product']).sum()
```
