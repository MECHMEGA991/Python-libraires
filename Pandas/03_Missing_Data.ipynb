##### Finding Missing Value

```python
import numpy as np
import pandas as pd
```

```python
data={
    'A':[1,2,np.nan,4,5],
    'B':[1,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5],
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum() # Total null value in every rows
```

```python
df.isna().any() # Is there any column where null value not
```

##### Removing Missong Data

```python
df
```

```python
df.dropna() # It will remove null value rows
```

```python
df.dropna(thresh=3) # drop rows where at least thresh value non-null value present
```

##### Filling Null Value

```python
df
```

```python
df.fillna(0) 
```

```python
# Using customize value:
val={'A':0,'C':15,'D':20} # Filling value will be columns wise
df.fillna(val)
```

```python
# Filling value with mean
df.mean()
```

```python
df.fillna(df.mean())
```
