##### Group by

```python
# import numpy as np
import pandas as pd
```

```python
data = {

    'Category' : ['A','B','A','B','A','B','A','B'],
    'Store'    : ['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales'    : [100,200,150,250,120,180,200,300],
    'Quantity' : [10,15,12,18,8,20,15,25],
    'Date'     : pd.date_range('2026-01-01',periods=8)

}
df=pd.DataFrame(data)
```

```python
# Group by Category and calculate the sum of sales
df.groupby('Category') # it returns a object u cant print that 

cat=df.groupby('Category')

for i,v in cat:
    print(i)
    print(v)
```

```python
cat=df.groupby('Category')['Sales'].sum()
cat
```

```python
# Group by Store and calculate the sum of sales

st=df.groupby('Store')['Sales'].sum()
st
```

```python
# Group by multiple columns 
# Group by Category and store 

st=df.groupby(['Category','Store'])['Sales'].sum()
st
```

#### Aggregation

```python
print(df['Sales'].mean())
```

```python
print(df['Sales'].median())
print(df['Sales'].std())
print(df['Sales'].var())
print(df['Sales'].min())
print(df['Sales'].max())
```

Aggregation keyword

```python
df['Sales'].agg(['mean','median','std','var','max','min','count'])
```
