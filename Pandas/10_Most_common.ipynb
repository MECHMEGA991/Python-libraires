### Grouping Data

```python

df1 = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "HR", "Sales","Sales"],
    "Employee": ["A", "B", "C", "D", "E","E"],
    "Salary": [50000, 40000, 60000, 45000, 55000,55000]
})
```

```python
df1
```

```python
df1.groupby('Department')['Salary'].max()
```

```python
df1.groupby('Department')['Salary'].agg(['max','mean'])
```

```python
df1['Difference'] = df1['Salary'] - df1.groupby('Department')['Salary'].transform('mean')
```

```python
df1
```
