###### A DataFrame in pandas is a 2-dimensional table-like data structure used to store and work with data in rows and columns.

```python
import numpy as np
import pandas as pd

# Creating a Dataframe
```

```python
data = {
    'Name' : ['John','Anna','Peter','Linda'],
    'Age'  : [20,34,29,42],
    'City' : ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000,70000,62000,85000]
}

df=pd.DataFrame(data)
```

```python
df
```

```python
# Using list

data_list=[
    ['John',20,'New York',65000],
    ['Anna',34,'Paris',62000],
    ['Peter',29,'Berlin',62000],
    ['Linda',42,'London',85000]
]

col=['Name','Age','City','Salary']
df2=pd.DataFrame(data_list,columns=col)
```
