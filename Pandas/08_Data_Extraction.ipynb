```python
# make a new column for episode count
df.loc[1]['Title']
```

```python
def extract_episode(txt):
    check = False
    data=""
    for i in txt:
        if i == ')':
            check=False
            return data
        if check==True:
            data = data +i
        if i== '(':
            check=True
            
        
    return data
```

```python
df['Episode']=df['Title'].apply(extract_episode)
```

```python
df['Episode']=df['Episode'].str.replace(" eps","")
```

```python
df['Episode']=df['Episode'].astype(int)
```
