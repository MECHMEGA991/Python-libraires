##### Pandas Series

###### A Series is a one-dimesional labeled array capable of holding any data type. The axix labels are collectively called the index

#### Creating Series
###### There are multiple ways to create a panda Series:

```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_list=[10,20,30]
arr=np.array([10,20,30])
d = {1:10,2:20,3:30}
```

```python
pd.Series(my_list,labels,name='temps')
```

```python
pd.Series(my_list) # Using list 
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr,index=labels) # Using numpy array
```

```python
pd.Series(d) # Using dictionaries
```
