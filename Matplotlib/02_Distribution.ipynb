## Distribution Plot


```python
'''
We use distribution plots to visualize the distribution of quantitative data

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are: 

 -> distplot/histplot
 -> jointplot
 -> pairplot
 -> rugplot
 

'''
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
```

```python
df['size'].unique() # This column is cosidered as categorical
```

```python
sns.displot(df['total_bill'])
```

```python
# Using Bins
sns.displot(df['total_bill'],bins=20)
```

```python
# Kernel density escumator 

sns.histplot(df['total_bill'],bins=20,kde=True) # Kde is used to show the bell curve
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],kde=True)
```

```python
# Join Plot For comparing two graphs 

sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
# pairplot
sns.pairplot(df)
```

```python
# hue => use to divide categorized by
sns.pairplot(df,hue='sex')
```

```python
# palette
sns.pairplot(df,hue='sex',palette='rainbow',kind='reg')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
# rugplot

sns.rugplot(df)
```

```python
sns.rugplot(df['total_bill'])
```
