```python
plt.subplot(2,2,1) # 2,2 is row into column for canvas
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,x)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y**3)
```

## Object Oriented Way

#### Matplotlib allows the aspect ratio, DPI and figure size to be specified when the figure is created. You can use the figsize and dpi keyword arguments

#### figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch)

```python
fig=plt.figure(figsize=(10,8),dpi=500)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)

axis2= fig.add_axes([0.4,0.6,0.3,0.3])
axis2.plot(y,x)
```
