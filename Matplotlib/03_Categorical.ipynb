### Categorical Data plots

```python
'''

Now let's discuss using seaborn to plot catrgorical data! There are few main plot types for this:

 -> boxplot
 -> violinplot
 -> stripplot
 -> barplot
 -> countplot

Let's go through example of each

'''
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.countplot(x=df['sex'])
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

```python
# In bar plot 
# -> y axis consists of qunatitative value
# -> x axis consists of categorical data
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'])
```

```python
'''
You can change the estimator object to your own function, that converts a vector to a scaler

Mean(default) : np.mean
Median        : np.median
Sum           : np.sum
Count         : len(to count the no. of observation)
SD            : np.std
Minimum       : np.min
Maximum       : np.max

'''
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```

```python

sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

## Boxplot and Violinplot

#### Boxplot and violinplot are used to shown the distribution of categorical data. 
#### A box plot (or box-and-whisker plot) shows the distribution of quantitatibe data in a way that facilitates comparision btw variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of distribution, except for points that are determined to be 'outliers' using a method that is a function of the inter-quartile range

```python
sns.violinplot(x='day',y='tip',data=df)
```

## Stripplot and swarmplot

The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drwan its own, but it is also a good complement to be a box or violin plot in cases where you want to show all the observation along with some representation of underlying distribution.

The swarmplot is similar to stripplot(). but the points are adjusted (only along the categorical axis) so that they don't overlap. This gives a better represntation of the distribution of values, although it does not scale as well to large observations(both in terms of the ability to show all the points and in terms of the computation needed to arrange them)

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='tip',y='day',data=df)
sns.swarmplot(x='tip',y='day',data=df)
```
