```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Heat Map

flights=sns.load_dataset('flights')
```

```python
flights
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
# Coorelation :- It describs the relationship btw two variables and how they move in relation to each other
# If one variable changes and the other also changes in a predictable way, they are correlated

# Three Types of Correlation
# +ve -> Both variables move in the same direction , When one increase, the other also increases. Study Time > --:> Exam score >

# -ve -> Variables move in opposite directions, When one increase, the other decreases. , price > --:> demand <

# zero(No) Correlation :> There is no relationship btw the variables, Shoe size and intelligence
```

```python
tipscorr=tips[['total_bill','tip','size']]
```

```python
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr())
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tipscorr)
```

## Cluster Map

```python
sns.clustermap(tipscorr.corr())
```
